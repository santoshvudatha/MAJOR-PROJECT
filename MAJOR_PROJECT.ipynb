```python
df['y'] = df['y'].fillna(df['y'].mean())
```

```python
#creating another column in the dataframe
df['Year'] = df['ds'].dt.year
```

```python
#forecasting : fbprophet - similar to regression
#rename for afghanisthan
df = df.rename(columns={'date' : 'ds' , 'total_cases' : 'y'})
```

```python
#creating 2 dataframes by groupy method
df_afghan = df[df['location'] == 'Afghanistan']
#df_afghan_avg = df_afghan.groupby('year').aggregate({'total_cases' : 'mean', 'total_deaths' : 'mean', 'total_vaccinations' : 'mean','people_vaccinated' : 'mean'})
df_afghan
```

```python
#another data frame 
df_guyan = df[df['location'] == 'Guyana']
#df_guyan_avg = df_guyan.groupby('Year').aggregate({'total_cases' : 'mean', 'total_deaths' : 'mean', 'total_vaccinations' : 'mean',
 #                                                    'people_vaccinations' : 'mean'})
df_guyan
```

```python
#graph for afghanisthan region
import matplotlib.pyplot as plt
plt.figure(figsize=(20,5))
plt.plot(df_afghan['ds'],df_afghan['y'],linestyle=':')
plt.plot(df_afghan['ds'],df_afghan['y'].rolling(30).mean())
plt.show()
```

```python
#graph for guyana region
plt.figure(figsize=(20,5))
plt.plot(df_guyan['ds'],df_guyan['y'])
plt.show()
```

```python
from fbprophet import  Prophet
model = Prophet(changepoint_prior_scale=0.5)
model.fit(df_afghan)
forecast = model.make_future_dataframe(periods=30,freq='D') #predicting for next 30 days
forecast = model.predict(forecast)
```

```python
model.plot(forecast,figsize=(20,5));
```

```python
from fbprophet import  Prophet
model = Prophet(changepoint_prior_scale=0.9)
model.fit(df_guyan)
forecast = model.make_future_dataframe(periods=30,freq='D')
forecast = model.predict(forecast)
```

```python
model.plot(forecast,figsize=(20,5));
```
